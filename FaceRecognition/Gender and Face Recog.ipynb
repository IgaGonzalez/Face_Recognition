```python
# author: Arun Ponnusamy
# website: https://www.arunponnusamy.com

# face detection webcam example
# usage: python face_detection_webcam.py 

# import necessary packages
from matplotlib import pyplot as plt
import cvlib as cv
import cv2
import keras
from keras.models import load_model
from keras.preprocessing.image import img_to_array
import face_recognition
import numpy as np
classifier = load_model("C:/Users/Nacho/model_v6_23.hdf5")
face_classifier = cv2.CascadeClassifier("C:/Users/Nacho/Paso a paso Reconocimiento Facial/detectors/haarcascade_frontalface_default.xml")
emotion_dict= {'Angry': 0, 'Sad': 5, 'Neutral': 4, 'Disgust': 1, 'Surprise': 6, 'Fear': 2, 'Happy': 3}
class_labels = {v:k for k,v in emotion_dict.items()}

def face_detector(img):
     # Convert image to grayscale
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    if faces is (): 
        print("Aca malo")
        return (0,0,0,0), np.zeros((48,48), np.uint8), img
        

    for (x,y,w,h) in faces:
        #cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        print("Aca bueno")
        roi_gray = gray[y:y+h, x:x+w]

        try:
            roi_gray = cv2.resize(roi_gray, (48, 48), interpolation = cv2.INTER_AREA)
            print("Ultra bueno")
        except:
            print("No tan bueno")
            return (x,w,y,h), np.zeros((48,48), np.uint8), img
        return (x,w,y,h), roi_gray, img
# open webcam
webcam = cv2.VideoCapture(0)



if not webcam.isOpened():
    print("Could not open webcam")
    exit()
    
process_this_frame = True

# loop through frames
while webcam.isOpened():
    try:
         # read frame from webcam 
        status, frame = webcam.read()
        
        if process_this_frame:  
            rect, cara, image = face_detector(frame)
            if np.sum([cara]) != 0.0:
                roi = cara.astype("float") / 255.0
                roi = img_to_array(roi)
                roi = np.expand_dims(roi, axis=0)

                
            predicted_class = np.argmax(classifier.predict(roi)[0])
             
            predicted_label = class_labels[predicted_class]
            print(predicted_class)
            pass
        process_this_frame = not process_this_frame
        
        if not status:
            print("Could not read frame")
            exit()

        # apply face detection
        face, confidence = cv.detect_face(frame)

        # loop through detected faces
        for idx, f in enumerate(face):

            (startX, startY) = f[0], f[1]
            (endX, endY) = f[2], f[3]

            # draw rectangle over face
            cv2.rectangle(frame, (startX,startY), (endX,endY), (0,255,0), 2)

            face_crop = np.copy(frame[startY:endY,startX:endX])
            (label,confidence) = cv.detect_gender(face_crop)

            idx = np.argmax(confidence)
            label = label[idx]

            Y = startY - 10 if startY - 10 > 10 else startY + 10
            
        
        
        
      # if process_this_frame:          
      #     
      #     if np.sum([face]) !=0.0:
      #         roi = face.astype("float") / 255.0
      #         roi = img_to_array(roi)
      #         roi = np.expand_dims(roi, axis=0)
      #         
      #         #preds = classifier.predict(roi)[0]
      #         predicted_class = np.argmax(classifier.predict(roi))
      #         predicted_label = class_labels[predicted_class]
      #         print(predicted_label)
      #         
        
        
        
        
        
         
        label = "{}:  {}: {:.2f}%".format(predicted_label,label,confidence[idx]*100)
            # write confidence percentage on top of face rectangle
        cv2.putText(frame, label, (startX,Y), cv2.FONT_HERSHEY_SIMPLEX, 0.7,
                       (0,255,0), 2)
        

 
    except Exception as e:
        print(e)
   
    # display output
    cv2.imshow("Real-time face detection", frame)

    # press "Q" to stop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
# release resources
webcam.release()
cv2.destroyAllWindows()     
```
